```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
# Replace 'R4 EFA 1st Combination.csv' with your local file path if needed
file_path = "/Users/stageacomeback/Desktop/Gerard Lee/PolyU SPEED RA/Exploratory Factor Analysis/Sumarised Dataset/Round 4 CSV/R4 EFA 1st Combination.csv"
data = pd.read_csv(file_path)

# Compute the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)

# Add title and labels
plt.title("Round 4 Combination 1", fontsize=16)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10, rotation=0)

# Show the plot
plt.tight_layout()
plt.show()
```
